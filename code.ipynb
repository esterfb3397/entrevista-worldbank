##### Question 5.2: A client works with files with a different format. They want us to implement that format in our dataset for the data from the "WesternEurope" region. The format they request is the following, sorted by country and metric (ascending order):
| Country           | Metric      | Value         | 
|-------------------|-------------|---------------|

```python
#First we create the new dataset
western_europe = continent[continent["region"] == "Western Europe"]
```

```python
#Then we drop the unncessary columns for this exercise and the repeated ones
western_europe.drop_duplicates(subset = "country", inplace = True)
western_europe = western_europe.drop(['Region',"region","code"], axis=1)
```

```python
def question52(df: pd.DataFrame):
    country: List[str] = df["country"].tolist()
    metric: List[str] = df.columns.tolist()
    df_copy = df.set_index("country")
    metric.remove("country")
    acc = []
    for c in country:
        for m in metric:
            acc.append([
                c, m, df_copy.at[c,m]
            ])
    return pd.DataFrame(acc, columns = ["Country", "Metric", "Value"])
```

```python
question52(western_europe)
```

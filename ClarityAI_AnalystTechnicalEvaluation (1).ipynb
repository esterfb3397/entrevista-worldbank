##### Question 2.2: A client would like to deal with fixed broadband subscription (per 100) and fixed telephone subscriptions (per 100) data as integers. They would also prefer the values be rounded down.

```python
#Fixed broadband subscription (per 100)
#First step,localice the column with the information
df.iloc[:,37]
#We round the numbers that are float
df.iloc[:,37] = pd.Series([round(val,2)for val in df.iloc[:,37]])
#Last step, we change the value
df.iloc[:,37] = df.iloc[:,37].astype(int)
```

```python
#Fixed telephone subscriptions(per 100)
#We localise the column
df.iloc[:,39]
#We round the numbers that are float
df.iloc[:,39] = pd.Series([round(val,2)for val in df.iloc[:,39]])
#Last step, we change the value
df.iloc[:,39] = df.iloc[:,39].astype(int)
```
